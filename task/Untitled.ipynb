```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

## Task 1: Load and Split the Dataset

Your task: Load the Iris dataset and split it into training and testing sets.


```python
# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Task 2: Explore the Dataset

Your Task: Print dataset (iris) parameters: feature names, target names, and the shape of the data.<br>
Replace '---' with appropriate code to print feature names, target names, and data shape.


```python
# Print feature names, target names, and data shape
print("Feature names:", iris.feature_names)
print("Target names:", iris.target_names)
print("Training data shape:", X_train.shape)
print("Test data shape:", X_test.shape)
```

## Task 3: Build and Train the TensorFlow Sequential Model

Your Task: Build a Sequential model with TensorFlow and train it on the Iris dataset.


```python
import tensorflow as tf
import numpy as np

# Convert labels to one-hot encoding
y_train_one_hot = tf.keras.utils.to_categorical(y_train, num_classes=3)
y_test_one_hot = tf.keras.utils.to_categorical(y_test, num_classes=3)

# Initialize the Sequential model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=3, activation='softmax', input_shape=(X.shape[1],))
])

# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train_one_hot, epochs=450)
```

## Task 4: Evaluate the Model

Your Task: Evaluate the trained model on the test dataset and print the accuracy and confusion matrix.


```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Make predictions on the test set
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```
